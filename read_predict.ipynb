```python
import numpy as np  
df = results

# 定义所有指标列
regression_cols = ['mse', 'mae', 'pearson_delta', 'mse_delta', 'mae_delta']
de_ranking_cols = ['pr_auc', 'roc_auc', 'overlap_at_100', 'precision_at_100', 'de_sig_genes_recall']
stats_cols = ['de_spearman_sig', 'de_spearman_lfc_sig', 'de_direction_match']
discrim_cols = ['discrimination_score_l1', 'discrimination_score_l2', 'discrimination_score_cosine']

# 计算并输出所有指标的平均值
def plot_metrics_summary(df, cols, title_map, filename, facecolor, n_rows, n_cols):
    """生成并保存箱线图"""
    # 根据 n_rows 和 n_cols 动态调整图表大小
    fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(5 * n_cols, 4 * n_rows))
    fig.suptitle(f'{title_map["main_title"]}分布 (Per-Perturbation)', fontsize=16)

    # 处理只有一个子图的情况
    if n_rows * n_cols == 1:
        axes = np.array([axes])
    axes = axes.flatten()
    
    # 填充平均值
    mean_metrics = df[cols].mean()

    for i, col in enumerate(cols):
        ax = axes[i]
        
        # 创建箱线图 (水平显示)
        ax.boxplot(df[col], vert=False, patch_artist=True, boxprops=dict(facecolor=facecolor))
        
        # 设置标题和标签
        ax.set_title(title_map[col], fontsize=12)
        ax.set_yticklabels([col])
        ax.grid(axis='x', linestyle='--', alpha=0.6)
        
        # --- 添加理想值和基线 ---
        ideal_val = 1.0
        baseline_val = 0.0
        
        # 确定基线值：0.5 用于 AUC、匹配度和判别分数；0.0 用于误差和相关性
        is_auc_or_match = col in ['roc_auc', 'pr_auc', 'de_direction_match', 
                                   'discrimination_score_l1', 'discrimination_score_l2', 
                                   'discrimination_score_cosine']
        
        if is_auc_or_match:
            baseline_val = 0.5
        
        # 绘制理想值线 (蓝色实线)
        ax.axvline(ideal_val, color='blue', linestyle='-', linewidth=1.5, label='理想值 (1.0)')
        
        # 绘制基线 (红色虚线)
        ax.axvline(baseline_val, color='red', linestyle='--', linewidth=1.5, label=f'基线 ({baseline_val})')
            
        # 标记平均值 (绿色圆点)
        mean_val = mean_metrics[col]
        ax.plot(mean_val, 1, 'o', color='darkgreen', markersize=8, label=f'Mean: {mean_val:.4f}')
        ax.legend(loc='upper right', fontsize=8)

    # 移除多余的子图
    for j in range(len(cols), len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    save_path = os.path.join(fig_save_path, filename)
    plt.savefig(save_path)
    plt.show()
    # plt.close()
    # print(f"已生成并保存图片: {filename}")


# 图表1:预测精度（回归）指标 (3 rows, 2 cols)
regression_titles = {
    'main_title': '1. 预测精度（回归）指标',
    'mse': '均方误差 (MSE)\n(越低越好, 基线: 0)',
    'mae': '平均绝对误差 (MAE)\n(越低越好, 基线: 0)',
    'pearson_delta': '皮尔逊相关性提升 ($\Delta$)\n(越高越好, 基线: 0)',
    'mse_delta': '均方误差提升 ($\Delta$ MSE)\n(越低越好, 基线: 0)',
    'mae_delta': '平均绝对误差提升 ($\Delta$ MAE)\n(越低越好, 基线: 0)'
}
plot_metrics_summary(df, regression_cols, regression_titles, 'plot_1_regression_metrics.png', 'lightblue', n_rows=3, n_cols=2)


# 图表2:DE 排名与重叠指标(3 rows, 2 cols)
de_ranking_titles = {
    'main_title': '2. 差异表达 (DE) 排名与重叠指标',
    'pr_auc': 'PR 曲线下面积 (PR-AUC)\n(越高越好, 随机基线: 0.5)',
    'roc_auc': 'ROC 曲线下面积 (ROC-AUC)\n(越高越好, 随机基线: 0.5)',
    'overlap_at_100': 'Top 100 重叠度\n(越高越好, 理想值: 1.0)',
    'precision_at_100': 'Top 100 精确率\n(越高越好, 理想值: 1.0)',
    'de_sig_genes_recall': '显著基因召回率 (Recall)\n(越高越好, 理想值: 1.0)'
}
plot_metrics_summary(df, de_ranking_cols, de_ranking_titles, 'plot_2_de_ranking_metrics.png', 'lightcoral', n_rows=3, n_cols=2)


# 图表3:统计相关性与方向指标（1 row, 3 cols）
stats_titles = {
    'main_title': '3. 统计相关性与方向指标',
    'de_spearman_sig': '显著基因的 Spearman 相关性\n(越高越好, 随机基线: 0.0)',
    'de_spearman_lfc_sig': '显著基因的 LFC Spearman 相关性\n(越高越好, 随机基线: 0.0)',
    'de_direction_match': '差异表达方向匹配度\n(越高越好, 随机基线: 0.5)'
}
plot_metrics_summary(df, stats_cols, stats_titles, 'plot_3_stats_metrics.png', 'lightsalmon', n_rows=1, n_cols=3)


# 图表4:分类与判别指标 (1 row, 3 cols)
discrim_titles = {
    'main_title': '4. 分类与判别指标',
    'discrimination_score_l1': 'L1 判别分数\n(越高越好, 随机基线: 0.5)',
    'discrimination_score_l2': 'L2 判别分数\n(越高越好, 随机基线: 0.5)',
    'discrimination_score_cosine': 'Cosine 判别分数\n(越高越好, 随机基线: 0.5)'
}
plot_metrics_summary(df, discrim_cols, discrim_titles, 'plot_4_discrim_metrics.png', 'lightgreen', n_rows=1, n_cols=3)
```

```python
from scipy.stats import spearmanr
from io import StringIO
N = 100 # Top N 基因数量
FDR_THRESHOLD = 0.05
TRUE_LFC_THRESHOLD = 0.58 # 对应 Fold Change 的 1.5 倍

# 筛选和合并目标基因数据

# 筛选扰动的数据
mbtps1_pred_df = predict_df[predict_df['target'].str.contains(PERTURBATION, case=False, na=False)].copy()
mbtps1_real_df = real_df[real_df['target'].str.contains(PERTURBATION, case=False, na=False)].copy()
# mbtps1_pred_df = predict_df[predict_df['reference'].str.contains(PERTURBATION, case=False, na=False)].copy()
# mbtps1_real_df = real_df[real_df['reference'].str.contains(PERTURBATION, case=False, na=False)].copy()

if mbtps1_pred_df.empty or mbtps1_real_df.empty:
    print(f"错误: 未找到 {PERTURBATION} 扰动的预测数据或真实数据。")

# 重命名关键列以区分预测值和真实值
mbtps1_real_df = mbtps1_real_df.rename(columns={
    'fold_change': 'true_fold_change',
    'fdr': 'true_fdr',
})
mbtps1_pred_df = mbtps1_pred_df.rename(columns={
    'fold_change': 'pred_fold_change',
    'fdr': 'pred_fdr',
})

# 合并预测值和真实值 (基于 feature/基因 ID)
analysis_df = pd.merge(
    mbtps1_pred_df[['feature', 'pred_fold_change', 'pred_fdr']],
    mbtps1_real_df[['feature', 'true_fold_change', 'true_fdr']],
    on='feature',
    how='inner'
)

# 计算 Log2 Fold Change (LFC)
# 检查 Fold Change 是否为非负，避免 log2(负数) 错误
analysis_df['true_lfc'] = np.log2(analysis_df['true_fold_change'].clip(lower=1e-6))
analysis_df['pred_lfc'] = np.log2(analysis_df['pred_fold_change'].clip(lower=1e-6))

squared_errors = (analysis_df['true_lfc'] - analysis_df['pred_lfc']) **2
mse = np.mean(squared_errors)
print(f"MSE of LFC prediction for {PERTURBATION}: {mse}")

merged_df = pd.merge(
    pred_de, 
    real_de, 
    on=["target", "feature"],  # 按目标基因和特征基因关联
    suffixes=("_pred", "_real")  # 区分预测和真实的列
)

# 对 fold_change 做对数转换（避免极端值影响，符合生信分析习惯）
merged_df["log2_fc_pred"] = np.log2(merged_df["fold_change_pred"].clip(lower=1e-6))
merged_df["log2_fc_real"] = np.log2(merged_df["fold_change_real"].clip(lower=1e-6))

# 计算 MSE（基于对数转换后的差异倍数）
mse = mean_squared_error(merged_df["log2_fc_real"], merged_df["log2_fc_pred"])

print(f"MSE between predicted and real log2(fold change): {mse:.4f}")

# 预测效果分析（三大维度）

# 回归准确性 (变化幅度)
spearman_corr, _ = spearmanr(analysis_df['true_lfc'], analysis_df['pred_lfc'])
mae = np.mean(np.abs(analysis_df['true_lfc'] - analysis_df['pred_lfc']))

# 分类准确性 (Top基因重叠)
# 识别真实 Top N 基因 (按 LFC 绝对值排序)
true_top_genes = analysis_df.sort_values(by='true_lfc', key=abs, ascending=False).head(N)['feature'].tolist()
# 识别预测 Top N 基因 (按 LFC 绝对值排序)
pred_top_genes = analysis_df.sort_values(by='pred_lfc', key=abs, ascending=False).head(N)['feature'].tolist()

# 计算重叠数和重叠率
overlap_count = len(set(true_top_genes) & set(pred_top_genes))
overlap_rate = overlap_count / N

# 识别真实显著基因数量
analysis_df['is_true_sig'] = (
    (analysis_df['true_fdr'] < FDR_THRESHOLD) & 
    (analysis_df['true_lfc'].abs() >= TRUE_LFC_THRESHOLD)
)
true_sig_count = analysis_df['is_true_sig'].sum()


# 打印结果并可视化

print("=" * 50)
print(f"{PERTURBATION} 基因扰动预测效果分析 ({len(analysis_df)} 个基因)")
print("=" * 50)

print("--- 1. 真实显著基因统计 ---")
print(f"  LFC >= ±{TRUE_LFC_THRESHOLD:.2f} 且 FDR < {FDR_THRESHOLD} 的真实显著基因数量: {true_sig_count}")
print("-" * 50)

print("--- 2. 回归准确性指标 (变化幅度) ---")
print(f"  Spearman 相关性 (LFC 排序): {spearman_corr:.4f} (理想值: 1.0)")
print(f"  平均绝对误差 (MAE, Log2FC): {mae:.4f} (理想值: 0.0)")
print("-" * 50)

print(f"--- 3. Top {N} 基因重叠分析 ---")
print(f"  真实 Top {N} 基因与预测 Top {N} 基因重叠数: {overlap_count}")
print(f"  重叠率: {overlap_rate:.4f} (理想值: 1.0)")
print("-" * 50)

# 可视化 LFC 比较 (散点图)
plt.figure(figsize=(9, 9))
plt.scatter(analysis_df['true_lfc'], analysis_df['pred_lfc'], s=10, alpha=0.6, color='darkorange')
plt.title(f'{PERTURBATION} 扰动: 预测 Log2FC vs. 真实 Log2FC\n(Spearman: {spearman_corr:.4f})', fontsize=14)
plt.xlabel('真实 Log2 Fold Change (True LFC)', fontsize=12)
plt.ylabel('预测 Log2 Fold Change (Predicted LFC)', fontsize=12)

# 绘制 y=x 对角线 (理想情况)
max_val = max(analysis_df['true_lfc'].abs().max(), analysis_df['pred_lfc'].abs().max()) + 0.5
limit = max(2.5, max_val) # 确保图表有足够的范围
plt.plot([-limit, limit], [-limit, limit], color='red', linestyle='--', linewidth=1.5, label='Ideal (y=x)')

# 绘制 LFC 显著性阈值线
plt.axhline(0, color='gray', linestyle='-', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='-', linewidth=0.5)

plt.xlim(-limit, limit)
plt.ylim(-limit, limit)
plt.legend(loc='upper left', fontsize=10)
plt.grid(True, linestyle=':', alpha=0.4)


plot_filename = f'{PERTURBATION}_LFC_prediction_scatter.png'

save_path = os.path.join(fig_save_path, plot_filename)
plt.savefig(save_path)

plt.show() # 仅为代码结构演示，不执行显示

# print(f"\n可视化代码结构已就绪，文件名为: {plot_filename}")
```

```python
# 比较分析real和pred之间宏观趋势和分布的匹配度，需要通过比较它们的统计特征，特别是扰动效应的幅度和显著性
df_compare = pd.DataFrame({
    'feature': real_df['feature'],
    'Real_FoldChange': real_df['fold_change'],
    'Pred_FoldChange': predict_df['fold_change'],
    'Real_PercentChange': real_df['percent_change'],
    'Pred_PercentChange': predict_df['percent_change'],
    'Real_FDR': real_df['fdr'],
    'Pred_FDR': predict_df['fdr']
}).dropna(subset=['Real_FoldChange', 'Pred_FoldChange'])
# 计算Spearman相关系数 (适用于排序和非线性关系)
spearman_corr = df_compare[['Real_FoldChange', 'Pred_FoldChange']].corr(method='spearman').iloc[0, 1]
print(f"Spearman 相关系数 (Fold Change): {spearman_corr:.4f}")

# 绘制 Fold Change 散点图
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x='Real_FoldChange', 
    y='Pred_FoldChange', 
    data=df_compare, 
    alpha=0.3,
    s=10 # 减小点的大小，因为有 18080 个点
)
# 绘制对角线 (y=x) 作为完美预测的基线
max_val = max(df_compare['Real_FoldChange'].max(), df_compare['Pred_FoldChange'].max())
min_val = min(df_compare['Real_FoldChange'].min(), df_compare['Pred_FoldChange'].min())
plt.plot([min_val, max_val], [min_val, max_val], 'r--', label='Perfect Prediction')

plt.xscale('log') # Fold Change 通常在 log 尺度上分析
plt.yscale('log')
plt.title(f'Real vs. Predicted Fold Change (Corr: {spearman_corr:.4f})')
plt.xlabel('Real Fold Change (Log Scale)')
plt.ylabel('Predicted Fold Change (Log Scale)')
plt.legend()
plt.grid(True, which="both", ls="--", linewidth=0.5)
plt.show()

# 定义真实和预测的显著差异基因
# 真实 DEG
real_deg_mask = (df_compare['Real_FDR'] < 0.05) & (np.abs(df_compare['Real_FoldChange']) > 1.2)
# 预测 DEG
pred_deg_mask = (df_compare['Pred_FDR'] < 0.05) & (np.abs(df_compare['Pred_FoldChange']) > 1.2)

real_deg_count = real_deg_mask.sum()
pred_deg_count = pred_deg_mask.sum()
correct_deg_count = (real_deg_mask & pred_deg_mask).sum()

# 计算精度 (Precision) 和召回率 (Recall)
# 精度: 模型预测的 DEG 中有多少是真正的 DEG
precision = correct_deg_count / pred_deg_count if pred_deg_count > 0 else 0
# 召回率: 真正的 DEG 中有多少被模型成功预测
recall = correct_deg_count / real_deg_count if real_deg_count > 0 else 0
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

print("\n--- 差异表达基因 (DEG) 匹配度 ---")
print(f"真实 DEG 数量: {real_deg_count}")
print(f"预测 DEG 数量: {pred_deg_count}")
print(f"正确预测的 DEG 数量 (交集): {correct_deg_count}")
print(f"DEG 预测精度 (Precision): {precision:.4f}")
print(f"DEG 预测召回率 (Recall): {recall:.4f}")
print(f"F1 Score: {f1_score:.4f}")
```
